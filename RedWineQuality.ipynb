```python
import statsmodels.api as sm
from patsy import dmatrices

y, X = dmatrices('quality ~ alcohol', data=wine, return_type='dataframe')
print("X:", type(X))
print(X.columns)

model = sm.OLS(y, X)  # OLS from statsmodels.api
result = model.fit()
print(result.summary())
```

```python
import statsmodels.formula.api as smf

model = smf.ols('quality ~ alcohol', data=wine)  # use ols, not OLS
results = model.fit()
print(results.params)
```

```python
wine['rate_code'] = (wine['quality'] > 4).astype(np.float32)
```

```python
y, X = dmatrices('rate_code ~ alcohol', data = wine)
sns.distplot(X[y[:,0] > 0, 1])
sns.distplot(X[y[:,0] == 0, 1])
```

```python
wine['binary_quality'] = (wine['quality'] >= 6).astype(int)

from patsy import dmatrices
import statsmodels.api as sm

# Use binary target variable
y, X = dmatrices('binary_quality ~ alcohol', data=wine, return_type='dataframe')

# Fit logistic regression
model = sm.Logit(y, X)
result = model.fit()

# Print summary
print(result.summary2())
```

```python
import seaborn as sns

yhat = result.predict(X)

sns.kdeplot(yhat[y.iloc[:, 0] > 0], label='Quality = 1')
sns.kdeplot(yhat[y.iloc[:, 0] == 0], label='Quality = 0')
```

```python
yhat = result.predict(X) > 0.955
print(sklearn.metrics.classification_report(y, yhat))
```

```python
model = sklearn.linear_model.LogisticRegression()
y,X = dmatrices('rate_code ~ alcohol + sulphates + citric_acid + fixed_acidity', data = wine)
model.fit(X, y)
yhat = model.predict(X)
print(sklearn.metrics.classification_report(y, yhat))
```

```python
y, X = dmatrices('rate_code ~ alcohol', data = wine)
model = sklearn.ensemble.RandomForestClassifier()
model.fit(X, y)
yhat = model.predict(X)
print(sklearn.metrics.classification_report(y, yhat))
```
