```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
# Constants
R = 2.0
a0 = 0.0						      #Start angle
x0 = 2.0; xc = x0				  #Start point x
y0 = 2.0; yc = y0				  #Start point y
arg = 6						      #Choose f(l, arg)
l0 = -0.5 if arg == 1 else 0.0
dl = 0.01
lmax = 2.0 * np.pi * R
```

```python
def f(l, arg):
    if arg == 1:
        return 1.0 / R
    elif arg == 2:
        return l / 2.0
    elif arg == 3:
        return math.log(l)
    elif arg == 4:
        return 0.5*l*l + 2*l + 1
    elif arg == 5:
        return 0.5*l*l*l + 2*l*l + 1
    else:
        return 0.5*l*l*l*l + 2*l*l*l + 4*l*l + 1
```

```python
l = np.linspace(0, 2.0*np.pi, 500)	#Truk of line
da = f(l, arg) * dl					      #Delta angel
a = a0 + da							          #Angels sum
x = x0 + dl * np.cos(a)				    #Convert polar truk to (x, y)
y = y0 + dl * np.sin(a)				    #Convert polar truk to (x, y)
xc += x; yc += y					        #Sum of (x, y) to (xc, tc)
```

```python
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title('Graph Plot arg = ' + str(arg),fontdict = font1)
plt.xlabel("-- Xc --", fontdict = font2)
plt.ylabel("-- Yc --", fontdict = font2)
plt.plot(xc, yc)
#Plot graphic# plt.show()
```
